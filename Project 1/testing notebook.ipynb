# Rainfall n stuff in Southeast China 
(mainly in the Fujian / Anhui regions)

```python
import logging
logging.basicConfig(level='INFO')

from ecmwf.datastores import Client
client = Client()
client.check_authentication()
```

```python
jobs = client.get_jobs(sortby="-created", status="successful")
request_ids = []
while jobs is not None:  # Loop over pages
    request_ids.extend(jobs.request_ids)
    jobs = jobs.next  # Move to the next page
```

```python
rqid = request_ids[0]

remote = client.get_remote(rqid)
```

```python
## extents

SE_CHINA_BBOX = [31.7, 115.0, 23.3, 123.2] ## covers fujian, zhejiang, southmost bit of anhui, bit of taiwan
```

```python
import earthkit as ek
import earthkit.data as ekd
import earthkit.plots as ekp

request = {
        'product_type': 'monthly_averaged_reanalysis',
        'variable': 'total_precipitation',
        'year': '2025',
        'month': '12',
        'time': '00:00',
        'data_format': 'netcdf',
        'area': SE_CHINA_BBOX
}

data = ekd.from_source(
    'cds', 'reanalysis-era5-single-levels-monthly-means', request
)
```

```python
data
```

```python
style = ekp.styles.Style(
    colors='Spectral_r',
    levels= range(232, 310, 2),
    units="kelvin",
    extend="both",
)


chart = ekp.Map(domain='China')

chart.grid_cells(data, style=style)

chart.coastlines(color="#555555")

chart.title("ERA5 monthly averaged {variable_name} - {time:%B %Y}")
chart.legend(label="{units}")
```
